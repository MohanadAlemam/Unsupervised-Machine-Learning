```python
## Exploratory Data Analysis:

import pandas as pd
data = pd.read_csv('Banknote_Authentication _Dataset.csv')
data
```

```python
## Calculations of Mean and Standard deviation eucladean distance :

import numpy as np
mean = np.mean(data,0)
std = np.std(data, 0)

## Also calculating maximum and minimum values:

max = np.max(data,0)
mini = np.min(data,0)

print (f'The colum-wise Means are: {mean} \n\nThe column-wise Standard Deviations are: {std}' )

print (f'\nThe maximum values are: {max} \n\nThe minimum values are {mini}')
```

```python
## Calculating the Euclidean distance of the mini and max of each feature:
data =  np.array(data)

euclidean_distance = [np.linalg.norm(point - mean) for point in data]

average_distance = sum(euclidean_distance) / len(euclidean_distance)

print (average_distance)
```

```python
## Running K-means 

from sklearn.cluster import KMeans
import numpy as np 

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Banknote_Authentication _Dataset.csv')

## The following np syntax tranfer the data into stack of values in a format of a list You can try printing ou thr outcome 'print(v1_v2)'

v1 = data['V1']
v2 = data['V2']

v1_v2 = np.column_stack((v1, v2))
km_res = KMeans(n_clusters=3, random_state=1000).fit(v1_v2)

## The (random_state=42) controls the starting point of the algorithm

clusters = km_res.cluster_centers_

plt.xlabel('Feature_1')
plt.ylabel('Feature_2')

plt.scatter(v1_v2[:, 0], v1_v2[:, 1], alpha=.23, s=30)

plt.scatter(clusters[:,0], clusters[:,1], s= 150,)
```
