```python
## Week Four


##1 Import the data file into Jypter using Upload button in the main window
##2 import pandas library using the following syntax:
import pandas as pd
##3 assign a variable name to your data eg (data) using the following syntax
data = pd.read_csv('happyscore_income.csv')
print (data)
```

```python
## We can acess a specific columns in the uploaded data using the following syntax:


happy = data ['happyScore']
income = data ['avg_income']
print (happy)
```

```python
## Now we can scattter the data 

import matplotlib.pyplot as plt
## We can label each aix of the plot using the following syntax:
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.scatter(income, happy)
```

```python
## We move on to sorting data by 'avarage_income'
data.sort_values('avg_income', inplace = True)

## If i want to filter for avg_income thats higher than a certain amount I use the following syntax: 
Richest = data[data['avg_income']>15000]

Richest
```

```python
## From the richest if we want to pull out a row eg whos the lowest income we can use the follwoing sytax

Richest.iloc[0]
```

```python
## From the richest if we want to pull out a row eg whos the highest income we can use the follwoing sytax

Richest.iloc[-1]
```

```python
# Now we can start doing some staticts eg finding the mean of the richest : TO DO SO WE IMPORT numpy

import numpy as np
Rich_mean= np.mean(Richest['avg_income'])
All_mean=np.mean(data['avg_income'])

print(Rich_mean, All_mean)
```

```python
### How to enahnce the plotting (Graphs Display) in the pandas

## we can plot the avge income and the Happy score for the richest, using the following sytax:

plt.scatter(Richest['avg_income'], Richest['happyScore'])

## not we want to add some text to the graph, this includes asssessing the position the text using x (avg income0 and y(happy_value) positioning 
plt.text(Richest.iloc[0]['avg_income'], 
         Richest.iloc[0]['happyScore'], 
         Richest.iloc[0]['country'])

## and for the last point we can use the following sytax
plt.text(Richest.iloc[-1]['avg_income'], 
         Richest.iloc[-1]['happyScore'], 
         Richest.iloc[-1]['country'])
```

```python
## What if we want to label all the points in the graph ? we can use the following code / syntax, We use the loop for:

plt.scatter(Richest['avg_income'], Richest['happyScore'])

for k, row in Richest.iterrows():
    plt.text(row['avg_income'], 
        row['happyScore'], 
         row['country'])
    
```

```python
## Now we want to interpret the data, to find link between the income and the happiness 
##1 Eyeball the data (look at it)
##2 What if i want to bring a 3rd dimension to the plot ie with happiness and income I want the colum 'income_inequality' to show.
## We can do that via making it appear as the size of the dot using the following syntax:

happy = data ['happyScore']
income = data ['avg_income']
ineq=data['income_inequality']

import matplotlib.pyplot as plt
plt.xlabel('Income')
plt.ylabel('Happiness Score')
plt.scatter(income, happy, s=ineq*8, alpha=.30)

## The s stands for size ie size of the dots, we time it by 6 to make it bigger, then added the alpha to manage transparency 

## X== Income , Y== Happiness , The size of the dots == the income inequality 
```

```python
### USING K-MEANS TO ANALYISE THE DATA
# The first thing is to import the K-mean from the 'sklearn.cluster':

from sklearn.cluster import KMeans

## Then we go about importing numpy

import numpy as np 

income_happy = np.column_stack((income, happy))
km_res = KMeans(n_clusters=3).fit(income_happy)

clusters = km_res.cluster_centers_
## Plot the data 
plt.scatter(income_happy[:, 0], income_happy[:, 1])

## Plot the K-means (centeriods):
plt.scatter(clusters[:,0], clusters[:,1], s= 600)
```
